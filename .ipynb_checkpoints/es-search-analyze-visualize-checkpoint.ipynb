```python
import json
import pandas as pd
from elasticsearch import Elasticsearch
%matplotlib inline
```

```python
es = Elasticsearch()
```

# Aggs

```python
query_string = {"query":{"bool":{"must":[{"match_all":{}}],"must_not":[],"should":[]}},"from":0,"size":0,"sort":[],"aggs":{"vendor":{"terms":{"field":"VENDOR_NAME.search","order":{"_count":"desc"},"size":10}}}}
```

```python
res = es.search(index = "icreqs", body=query_string)
```

```python
# the results are a list/array of dictionaries notice the [] brackets
res['aggregations']['vendor']['buckets']
```

```python
# convert list to datafram
df_aggs = pd.DataFrame.from_dict(res['aggregations']['vendor']['buckets'])
```

```python
df_aggs
```

### Pandas Vis

```python

df_aggs.plot.bar(y='doc_count', x='key')
```

## Seaborn Vis

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))
# reset fonts etc to default
sns.set()
g = sns.barplot(x=df_aggs.key, y=df_aggs.doc_count)
#rotate labels
labels = g.get_xticklabels()
g.set_xticklabels(labels,rotation=90)
```

# Data  Search only

```python
query_string = {"query":{"bool":{"must":[{"match_all":{}}],"must_not":[],"should":[]}},"from":0,"size":250,"sort":[],"aggs":{}}
```

```python
res = es.search(index = "icreqs", body=query_string)
```

```python
 # this is a list of dictionaries
res_list = res['hits']['hits']
res_list
```

```python
df_data_array = []
for x in res_list:
    print(x['_source'], '\n')
    df_data_array.append(x['_source'])
```

```python
df_data_array
```

```python
df_data = pd.DataFrame.from_dict(df_data_array)
```

```python
df_data.head()
```

# Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))
# reset fonts etc to default
sns.set()
g = sns.countplot(x=df_data.BUYER_ID, data=df_data)
#rotate labels
labels = g.get_xticklabels()
g.set_xticklabels(labels,rotation=90)
```
