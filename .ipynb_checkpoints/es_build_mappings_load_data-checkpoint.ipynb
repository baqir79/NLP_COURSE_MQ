```python
import pandas as pd
import json
```

```python
df = pd.read_excel('creqs.xlsx',sheetname='Sheet2')
df_data = pd.read_excel('creqs.xlsx',sheetname='Sheet1')
```

```python
#df_data
```

```python
#build mapping template
appname = 'creqs'
sindex = "i" + appname # the index
stype =   appname  # soucre doc type
mapj = {}
mapj['mappings'] = {}
mapj['mappings'][stype] =  {"properties": {}}
#mapj['mappings'][stype]['_all']
#mapj['mappings'][stype]['properties'] = {}

mbody = json.dumps(mapj)
print(mapj)
```

```python
for index, row in df.iterrows():
    # build jsons
    fmap = {}
    if row['type'] == 'date':
        # date
        fmap = {"null_value": "null", "format": "MM/dd/yy","ignore_malformed": "true","type": "date"}
    elif row['type'] == 'keytext':
        fmap = {"type": "keyword", "fields" : {"search" : {"type" : "text","fielddata": "true"}}}
    else:
        fmap = {"type": row['type']}
        
    mapj['mappings'][stype]['properties'][row['fname']] = fmap
```

```python
print(json.dumps(mapj['mappings']))
```

```python
from elasticsearch import Elasticsearch
# conntect es
es = Elasticsearch()
# delete index if exists
print (sindex)
if es.indices.exists(sindex):
    es.indices.delete(index=sindex)
# index settings load analyzers template
with open('analyzers.json') as f:
    analyzers = json.load(f)
    
#print(analyzers)
settings = {"settings": analyzers['settings'], "mappings": mapj['mappings']}
#print(settings)
# create index
es.indices.create(index=sindex, ignore=400, body=settings)
```

```python
# create a json dump from data frame
df_as_json = df_data.to_json(orient='records', lines=True)

# break on new line so that you can iterate by each record 
for doc in df_as_json.split('\n'):
    print(doc)
    res = es.index(index=sindex , doc_type=stype, body=doc)
```

# csv to json records builder

```python
import csv
import json

csvfile = open('file.csv', 'r')
jsonfile = open('file.json', 'w')

fieldnames = ("FirstName","LastName","IDNumber","Message")
reader = csv.DictReader( csvfile, fieldnames)
for row in reader:
    json.dump(row, jsonfile)
    jsonfile.write('\n')
```
