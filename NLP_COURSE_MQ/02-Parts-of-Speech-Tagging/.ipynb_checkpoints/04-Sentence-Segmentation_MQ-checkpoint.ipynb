```python
def set_custom_boundries(doc):
    for token in doc[:-1]:
        if token.text == ';':
            doc[token.i+1].is_sent_start = True
    return doc
```

```python
nlp.add_pipe(set_custom_boundries, before='parser')

nlp.pipe_names
```
